```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv("weather.csv")
```

```python
weather1 = weather[weather.Station == 1]
```

```python
weather1_c = weather1[["Date", "PrecipTotal"]]
```

```python
weather1_c = weather1_c.replace("T", 0.005, regex=True)
weather1_c = weather1_c.replace("M", np.nan, regex=True)
```

```python
weather1_c['MovingAverage'] = 0
```

```python
weather1_c['MovingAverage'] = pd.rolling_mean(weather1_c.PrecipTotal, window=21, min_periods=1)
```
