```python
tmp = []
for species in count_2.Species.values:
    for date in count_2.Date.values:
        mask_species = X.Species ==species
        mask_date = X.Date == date
        mask = mask_species & mask_date #all(mask_species, mask_date)
        mask_count_date = count_2.Date == date 
        mask_count_species = count_2.Species == species
        mask_count = mask_count_species & mask_count_date
        #print(mask)
        #print(len(mask_count))
        #print(len(count_2))
        #print(count_2[mask_count])
        X[species]= 0
        X[species][mask] = count_2.counter[mask_count]
        tmp.append(count_2.counter[mask_count])
    #X.ix[(X.Species == species, X.Date == date).all(),species] = -1
```
